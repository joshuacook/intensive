```python
adult_df = pd.read_pickle('data/adult_cluster.p')
```

```python
adult_df.columns = ['fit_time_gm_full', 'model_x', 'bic_gm_full', 'fit_time_gm_tied', 'model_y',
       'bic_gm_tied', 'fit_time_gm_diag', 'model_x', 'bic_gm_diag', 'fit_time_gm_spherical',
       'model_y', 'bic_gm_spherical', 'fit_time_km', 'model', 'bic_km']
```

```python
_, ax = plt.subplots(1,2,figsize=(20,8))
adult_df[adult_df.index < 35][['bic_km','bic_gm_full', 'bic_gm_tied', 'bic_gm_diag', 'bic_gm_spherical']].plot(ax=ax[0])
adult_df[adult_df.index < 35][['fit_time_km', 'fit_time_gm_full', 'fit_time_gm_tied',  'fit_time_gm_diag', 'fit_time_gm_spherical']].plot(ax=ax[1])
ax[0].set_title('BIC by number of clusters')
ax[1].set_title('Fit Time by number of clusters')

```

```python
madelon_df = pd.read_pickle('data/madelon_cluster.p')
```

```python
madelon_df.columns = ['fit_time_gm_full', 'model_x', 'bic_gm_full', 'fit_time_gm_tied', 'model_y',
       'bic_gm_tied', 'fit_time_gm_diag', 'model_x', 'bic_gm_diag', 'fit_time_gm_spherical',
       'model_y', 'bic_gm_spherical', 'fit_time_km', 'model', 'bic_km']
```

```python
_, ax = plt.subplots(1,4,figsize=(20,8))
adult_df[['bic_km', 'bic_gm_spherical']].plot(ax=ax[0])
adult_df[['fit_time_km', 'fit_time_gm_spherical']].plot(ax=ax[1])
ax[0].set_title('Adult BIC by number of clusters')
ax[1].set_title('Adult Fit Time by number of clusters')
madelon_df[['bic_km', 'bic_gm_spherical']].plot(ax=ax[2])
madelon_df[['fit_time_km', 'fit_time_gm_spherical']].plot(ax=ax[3])
ax[2].set_title('Madelon BIC by number of clusters')
ax[3].set_title('Madelon Fit Time by number of clusters')
```
